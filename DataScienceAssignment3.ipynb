```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
html_doc = """ 
<html>
<body>

<h2>HTML Table</h2>

<table>
	<tr>
		<th>Student</th>
		<th>Favourite Computer Language</th>
		<th>Years of Experience</th>
	</tr>
	<tr>
		<td>Fred</td>
		<td>Kotlin</td>
		<td>4</td>
	</tr>
	<tr>
		<td>Paula</td>
		<td>Python</td>
		<td>2</td>
	</tr>
	<tr>
		<td>Ernst</td>
		<td>Java</td>
		<td>7</td>
	</tr>
	<tr>
		<td>Isabel</td>
		<td>C++</td>
		<td>3</td>
	</tr>
	<tr>
		<td>Tony</td>
		<td>Pearl</td>
		<td>7</td>
	</tr>
	<tr>
		<td>Maria</td>
		<td>Cobol</td>
		<td>15</td>
	</tr>
</table>
""" 
soup = BeautifulSoup(html_doc, 'html.parser')
table = soup.find('table')
print(table)
```

```python
headers = table.find_all("th")
print(headers)
```

```python
titles = []

for i in headers:
    title = i.text
    titles.append(title)

print(titles)
```

```python
df = pd.DataFrame(columns = titles)
print(df)
```

```python
rows = table.find_all("tr")
print(rows)
```

```python
for i in rows[1:7]:
    print(i)
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]
    l = len(df)                        
    df.loc[l] = row

df.head(6)
```
